Υπολογισμός  στατιστικών ανά Stint ανά οδηγό. Stint = ο αριθμός των γύρων ου κάνει ένας οδηγος με τα ίδια ελαστικά, χωρίς να μπει pit stop. Ξεκινάει από την έξοδο και τελειώνει στην έισοδο στα pit.

Βήμα 1 — Έλεγχος αξιοπιστίας:

Αν το stint έχει λιγότερους από 4 γύρους, επιστρέφει None — δεν υπολογίζει τίποτα. Με λιγότερους από 4 γύρους η κλίση δεν είναι στατιστικά αξιόπιστη γιατί ένας μόνο outlier γύρος θα αλλοιώσει το αποτέλεσμα.

Βήμα 2 — Linear regression:

Χρησιμοποιεί το np.polyfit(TyreLife, LapTime, 1) για να βρει την ευθεία που "ταιριάζει" καλύτερα στα δεδομένα. Το TyreLife είναι ο άξονας x — πόσοι γύροι έχουν γίνει στο τρέχον σετ ελαστικών. Το LapTime είναι ο άξονας y — ο χρόνος γύρου σε δευτερόλεπτα. Το 1 σημαίνει ότι ψάχνει ευθεία γραμμή — όχι καμπύλη.

Βήμα 3 — Εξαγωγή κλίσης:

Το np.polyfit επιστρέφει δύο αριθμούς — κλίση και intercept. Το [0] κρατά μόνο την κλίση. Αυτός ο αριθμός λέει: "για κάθε επιπλέον γύρο στο ίδιο σετ ελαστικών, ο χρόνος ανεβαίνει κατά X δευτερόλεπτα".

Έξοδος: Ένας αριθμός — π.χ. 0.087 σημαίνει ότι ο οδηγός χάνει 0.087 δευτερόλεπτα ανά γύρο λόγω φθοράς ελαστικών.

```python
deg_rate_temp = r_laps.groupby(["Driver", "Stint"]).size()
degradation_rate = r_laps.groupby(["Driver", "Stint"]).count()
degradation_rate["Stint_Laps"] = degradation_rate["DriverNumber"]
r_laps[r_laps["Driver"] == "ALB"][["Driver", "Stint", "TyreLife", "LapTime"]].head(20)

```

```python
def calc_deg_rate(stint_df):
    stint_df = stint_df[stint_df["TyreLife"] > 1]  # αφαίρεσε πρώτο γύρο
    stint_df = stint_df[stint_df["Stint"] > 1]
    if len(stint_df) < 4:                           # έλεγξε αν έμειναν αρκετοί
        return None
    slope = np.polyfit(stint_df["TyreLife"], stint_df["LapTime"], 1)[0]
    return slope

degradation_rate = r_laps.groupby(["Driver", "Stint"]).apply(calc_deg_rate)
degradation_rate = pd.DataFrame(degradation_rate)
degradation_rate = degradation_rate.reset_index()
degradation_rate["Deg_Rate"] = degradation_rate[0]
degradation_rate = degradation_rate.drop([0], axis = 1)
degradation_rate
```

Weighted Average


```python
stint_sizes = r_laps.groupby(["Driver", "Stint"]).size().reset_index()
stint_sizes = stint_sizes.rename(columns={0: "Laps"})
merged_df = pd.merge(stint_sizes, degradation_rate, on=['Driver','Stint'], how='inner')
merged_df
```

```python
merged_df_clean = merged_df[merged_df["Deg_Rate"].notna()]

deg_rate_final = merged_df_clean.groupby("Driver").apply(
    lambda x: (x["Deg_Rate"] * x["Laps"]).sum() / x["Laps"].sum()
).reset_index()

deg_rate_final = deg_rate_final.rename(columns={0: "Deg_Rate_Weighted"})
deg_rate_final
```
